```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#matlab style formatting
plt.plot(x_axis, y_axis)
plt.xlabel("Month")
plt.ylabel("Total Fare")
plt.title("Fares By Month")
plt.show()
```

```python
# object-oriented formatting
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="o", color="blue", linewidth=2, label='Boston')

#create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#Set the y limit between 0 and 45
plt.ylim(0,45)

#Create a title
plt.title("PyBear Fare by Month")

#add a legend
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth="2", label='Boston')
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0,45)
ax.grid()
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```
